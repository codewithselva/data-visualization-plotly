```python
import pandas as pd
import os
import json
```

```python
path_agg_ins_ind = "C:/Users/malar/Documents/github_codewithselva/data-visualization-plotly/data/external/data/aggregated/insurance/country/india/"
list_year = os.listdir(path_agg_ins_ind)

dict_agg_ins_all = {'year': [], 'quarter':[], 'country': [], 'state': [], 'transaction_type': [], 'transaction_count': [], 'transaction_amount': []}

for year in list_year:
    if(year.isdigit()):
        path_year = path_agg_ins_ind + year + "/"
        file_list = os.listdir(path_year)
        for file in file_list:
            path_file = path_year + file
            data_file = open(path_file, 'r')
            data_json = json.load(data_file)
            for data in data_json['data']['transactionData']:
                transaction_type = data['name']
                transaction_count = data['paymentInstruments'][0]['count']
                transaction_amount = data['paymentInstruments'][0]['amount']
                dict_agg_ins_all['year'].append(year)
                dict_agg_ins_all['quarter'].append(int(file.strip('.json')))
                dict_agg_ins_all['country'].append('India')
                dict_agg_ins_all['state'].append(None)
                dict_agg_ins_all['transaction_type'].append(transaction_type)
                dict_agg_ins_all['transaction_count'].append(transaction_count)
                dict_agg_ins_all['transaction_amount'].append(transaction_amount)
                
df_agg_ins_all = pd.DataFrame(dict_agg_ins_all)
```
