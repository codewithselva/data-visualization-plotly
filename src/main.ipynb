```python
import os
import json
import pandas as pd
```

```python
state_path = "C:/Users/malar/Documents/github_codewithselva/data-visualization-plotly/src/resources/data/aggregated/transaction/country/india/state/"
states = os.listdir(state_path)
```

```python
state_info={"State":[], "Year":[], "Quarter":[],"Transaction_Type":[], "Transaction_Count":[], "Transaction_Amount":[]}

for state in states:
    path_state = state_path+state+'/'
    year_list = os.listdir(path_state)
    
    for year in year_list:
        curr_year = path_state+year+'/'
        file_list = os.listdir(curr_year)
        for file in file_list:
            curr_file = curr_year+file
            file_data = open(curr_file,'r')
            data_json = json.load(file_data)
            for data in data_json['data']['transactionData']:
                #print(data)
                name = data['name']
                count = data['paymentInstruments'][0]['count']
                amount = data['paymentInstruments'][0]['amount']
                #print(f"====> NAME: {name} ; COUNT: {count}; AMOUNT: {amount}")
                state_info['Transaction_Type'].append(name)
                state_info['Transaction_Count'].append(count)
                state_info['Transaction_Amount'].append(amount)
                state_info['State'].append(state)
                state_info['Year'].append(year)
                state_info['Quarter'].append(int(file.strip('.json')))
```

```python
df_transaction = pd.DataFrame(state_info)
```

```python
#User Dataframe
dir_states = "C:/Users/malar/Documents/github_codewithselva/data-visualization-plotly/src/resources/data/aggregated/user/country/india/state/"

states = os.listdir(dir_states)

device_info={"State":[], "Year":[], "Quarter":[],"Brand":[], "Count":[], "Percentage":[]}

for state in states:
    #print("STATE -> ",state)
    path_state = dir_states+state+'/'
    #print("path_state -> ",path_state)
    year_list = os.listdir(path_state)
    
    for year in year_list:
        #print("YEAR -> ",year)
        curr_year_dir = path_state+year+'/'
        #print("curr_year_dir -> ",curr_year_dir)
        file_list = os.listdir(curr_year_dir)
        for file in file_list:
            #print("FILE -> ", file)
            curr_file_dir = curr_year_dir+file
            file_data = open(curr_file_dir,'r')
            data_json = json.load(file_data)
            #print("data_json -> ",data_json)
            
            try:
                for data in data_json['data']['usersByDevice']:
                    #print(data)
                    brand = data['brand']
                    count = data['count']
                    percentage = data['percentage']
                    device_info['Brand'].append(brand)
                    device_info['Count'].append(count)
                    device_info['Percentage'].append(percentage)
                    device_info['State'].append(state)
                    device_info['Year'].append(year)
                    device_info['Quarter'].append(int(file.strip('.json')))
            except KeyError as ke:
                #print(ke)
                pass
            except:
                pass

```

```python
df_device_stats = pd.DataFrame(device_info)
```

```python
#Map Dataframe

dir_map_trans = "C:/Users/malar/Documents/github_codewithselva/data-visualization-plotly/src/resources/data/map/transaction/hover/country/india/state/"
map_trans = os.listdir(dir_map_trans)

map_trans_info={"State":[], "Year":[], "Quarter":[],"District":[], "Count":[], "Amount":[]}

for state in map_trans:
    #print("STATE -> ",state)
    path_state = dir_map_trans+state+'/'
    #print("path_state -> ",path_state)
    year_list = os.listdir(path_state)
    
    for year in year_list:
        #print("YEAR -> ",year)
        curr_year_dir = path_state+year+'/'
        #print("curr_year_dir -> ",curr_year_dir)
        file_list = os.listdir(curr_year_dir)
        for file in file_list:
            #print("FILE -> ", file)
            curr_file_dir = curr_year_dir+file
            file_data = open(curr_file_dir,'r')
            data_json = json.load(file_data)
            #print("data_json -> ",data_json)
            try:
                for data in data_json['data']['hoverDataList']:
                    #print(data)
                    district = data['name']
                    count = data['metric'][0]['count']
                    amount = data['metric'][0]['amount']
                    map_trans_info['District'].append(district)
                    map_trans_info['Count'].append(count)
                    map_trans_info['Amount'].append(amount)
                    map_trans_info['State'].append(state)
                    map_trans_info['Year'].append(year)
                    map_trans_info['Quarter'].append(int(file.strip('.json')))
            except KeyError as ke:
                #print(ke)
                pass
            except:
                pass
```

```python
dir_map_user = "C:/Users/malar/Documents/github_codewithselva/data-visualization-plotly/resources/data/map/user/hover/country/india/state/"

map_users = os.listdir(dir_map_user)

map_user_info={"State":[], "Year":[], "Quarter":[],"District":[], "Registered_Users":[], "AppOpens":[]}

for state in map_users:
    #print("STATE -> ",state)
    path_state = dir_map_user+state+'/'
    #print("path_state -> ",path_state)
    year_list = os.listdir(path_state)
    
    for year in year_list:
        #print("YEAR -> ",year)
        curr_year_dir = path_state+year+'/'
        #print("curr_year_dir -> ",curr_year_dir)
        file_list = os.listdir(curr_year_dir)
        for file in file_list:
            #print("FILE -> ", file)
            curr_file_dir = curr_year_dir+file
            file_data = open(curr_file_dir,'r')
            data_json = json.load(file_data)
            print("data_json -> ",data_json)
            try:
                for data in data_json['data']['hoverData'].items():
                    district = data[0]
                    registeredUser = data[1]['registeredUsers']
                    appOpens = data[1]['appOpens']
                    map_user_info['District'].append(district)
                    map_user_info['Registered_Users'].append(registeredUser)
                    map_user_info['AppOpens'].append(appOpens)
                    map_user_info['State'].append(state)
                    map_user_info['Year'].append(year)
                    map_user_info['Quarter'].append(int(file.strip('.json')))
            except KeyError as ke:
                #print(ke)
                pass
            except:
                pass
```
